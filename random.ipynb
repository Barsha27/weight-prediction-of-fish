```python
# finding the missing no in array

import collections
def finder(arr1, arr2):

    d = collections.defaultdict(int)

    for num in arr2:
        d[num] += 1

    print(d)

    for num in arr1:
        if d[num] == 0:
            print(num)

        else:
            d[num] -= 1
```

```python
arr1 = [10,3,5]
arr2 = [3,5]

finder(arr1,arr2)
```

```python
# reversing a word

def rev_word(s):

    s = s.lower()

    length = len(s)

    for i in range(length-1,-1,-1):
        print(s[i], end = "")

```

```python
s = "Barsha"
rev_word(s)
```

```python
# characters and print it in descending order of their count

def count_char(s):

    # making an empty dictionery

    char = dict()

    # counting the values of each character and appending it to dictionery

    for i in s:
        if s.count(i) > 1:
            c1 = s.count(i)
            char[c1] = i

    print(char)
    
    # sorting the value by keys and priting the corresponding key

    for i in sorted(char.keys(), reverse = True):

        print(char[i], end = "")

        

 
            
```

```python
s = "aaabbccccc"
count_char(s)
```
